```python
import pandas as pd
```

```python
train_df=pd.read_csv("train_data.txt", sep=":::")
train_df.columns=['index','movie_name', 'genre', 'plot']
train_df.drop(['index', 'movie_name'], axis=1, inplace=True)
train_df
```

```python
test_df=pd.read_csv("test_data.txt", sep=":::")
test_df.columns=['index','movie_name', 'plot']
test_df.drop(['index', 'movie_name'], axis=1, inplace=True)
test_df
```

```python
# preprocessing text data
train_df.isna().sum()
test_df.isna().sum()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer(stop_words='english', max_features=5000)
x_train=tf.fit_transform(train_df['plot'])
x_test=tf.transform(test_df['plot'])
```

```python
train_df['genre']
```

```python
from sklearn.preprocessing import LabelEncoder
lbl=LabelEncoder()
y_train=lbl.fit_transform(train_df['genre'])
y_train
```

```python
# model training
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB().fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
def input_fun(text):
    text_clean = tf.transform([text])
    y_pred_encoded = model.predict(text_clean)  # Use the transformed text
    y_pred_genre = lbl.inverse_transform(y_pred_encoded)  # Convert encoded labels back to genres
    print(y_pred_genre)
```
